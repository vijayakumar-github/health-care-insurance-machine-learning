```python
# create a new column for the charges category

def charges_category(charges):
    if charges < 20000:
        return 'high'
    else:
        return 'low'

df_numeric['charges_category'] = df_numeric['charges'].apply(charges_category)
```

```python
df_numeric
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Instance:

ordinal = OrdinalEncoder()

df_numeric['charges_category'] = ordinal.fit_transform(df_numeric[['charges_category']])

df_numeric.head()
```

```python
df_numeric.drop(['charges'],axis=1)
```

```python
x1=df_numeric[['age', 'bmi', 'children']]
y1=df_numeric['charges_category']
```

```python
# spliting the above data into train and test set split in the ratio of 80:20

x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=1)
```

```python
x1_train,x1_test,y1_train, y1_test
```

```python
model=LogisticRegression()
model.fit(x1_train, y1_train)
```

```python
y1_pred=model.predict(x1_test)
```

```python
print(y1_pred)
print(y1_test)
```

```python
#classification_report LIBRARY
from sklearn.metrics import classification_report

#classification_report
print(classification_report(y1_test,y1_pred))
```

```python
from sklearn import metrics 

conf_mat=metrics.confusion_matrix(y1_test, y1_pred)
print('Confusion Matrix : ', conf_mat)
print()

Accuracy_score=metrics.accuracy_score(y1_test, y1_pred) 
print('Accuracy Score : ', Accuracy_score) 
print()

print('Accuracy in Percentage : ', int(Accuracy_score*100),'%')
```
